```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(5)
```

## Player Count

* Display the total number of players


```python
player_count = pd.DataFrame({"count": purchase_data["SN"].nunique()},index = [1])
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
analysis = pd.DataFrame({"# of Unique Items": purchase_data["Item ID"].nunique(),
                        "Average Price": purchase_data["Price"].mean(),
                        "# of Purchases" : len(purchase_data),
                         "Total Revenue": purchase_data["Price"].sum()
                        },index = [0])
analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count = purchase_data.groupby(["Gender"])
gender_ct =gender_count["Gender"].count()
gender_pct = round(((gender_ct) / purchase_data["SN"].nunique()) * 100,2)

gender_summary = pd.DataFrame({"Total Count": gender_ct,
                              "Percentage of Players": gender_pct},index = ["Male","Female","Other / Non-Disclosed"])
gender_summary.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = gender_count["Purchase ID"].count()
purchase_avg_price = gender_count["Price"].mean().map("${:,.2f}".format)
purchase_total = gender_count["Price"].sum()
purchase_avg_person = (purchase_total / gender_count["SN"].nunique()).map("${:,.2f}".format)
gender_purchase = pd.DataFrame({"Purchase Count": purchase_count,
                             "Average Purchase Price": purchase_avg_price,
                               "Total Purchase Value": purchase_total.map("${:,.2f}".format),
                               "Avg Total Purchase per Person": purchase_avg_person},
                               index = ["Male","Female","Other / Non-Disclosed"])
gender_purchase.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,10,15,20,25,30,35,40,100]
bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["Age Bucket"] = pd.cut(purchase_data["Age"], bins, labels=bin_names)
purchase_data.head()
age_purchase = purchase_data.groupby(["Age Bucket"])
age_purchase.count().head()
age_count = age_purchase["Age Bucket"].count()
age_pct = round((age_count / sum(age_count) * 100),2)
#age_count
#age_pct
age_summary_table = pd.DataFrame({"Total Number": age_count, "Percentage": age_pct})
age_summary_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# purchase count, avg purchase price, total purchase value, avg total purchase per person
age_purchase.count().head()
purchase_count = age_purchase["Price"].count()
purchase_total = age_purchase["Price"].sum()
purchase_pct = round((purchase_count / purchase_total * 100),2)
purchase_avg = purchase_total / purchase_count
#purchase_avg
purchase_summary_table = pd.DataFrame({"Purchase Count": purchase_count, 
                                        "Avg Purchase Price": purchase_pct.map("${:,.2f}".format),
                                        "Total Purchase Value": purchase_total.map("${:,.2f}".format),
                                        "Avg Total Purchase per Person": purchase_avg.map("${:,.2f}".format)
                                      })
purchase_summary_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#SN, purchase count, avg purchase price, total purchase value
top_spend = purchase_data[["SN","Price","Purchase ID"]]
top_spend_SN = top_spend.groupby(["SN"])
spend_count = top_spend_SN["SN"].count()
value = top_spend_SN["Price"].sum()
avg_spend = value / spend_count
spend_summary = pd.DataFrame({"Purchase Count": spend_count, 
                                    "Avg Purchase Price": avg_spend.map("${:,.2f}".format),
                                   "Total Purchase Value": value.map("${:,.2f}".format)})
spend_summary = spend_summary.sort_values("Total Purchase Value", ascending=False)
spend_summary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# purchase count, item price, total purchase value
most_pop = purchase_data[["Item ID","Item Name","Price","Purchase ID"]]
most_pop_group = most_pop.groupby(['Item ID', 'Item Name'])
most_pop_group.count().head()
pop_count = most_pop_group["Purchase ID"].count()
tot_price = most_pop_group["Price"].sum()
price = tot_price / pop_count

most_pop_table = pd.DataFrame({"Purchase Count":pop_count,
                              "Item Price" : price.map("${:,.2f}".format),
                              "Purchase Value" : tot_price.map("${:,.2f}".format)})
most_pop_table = most_pop_table.sort_values("Purchase Count",ascending =False)
most_pop_table.head()
```
